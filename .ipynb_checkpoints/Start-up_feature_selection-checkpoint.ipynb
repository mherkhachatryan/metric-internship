# Feature Importance Selection 

## Training

```python
from sklearn.metrics import r2_score,make_scorer,f1_score, mean_squared_error, accuracy_score, confusion_matrix, precision_score, recall_score
from sklearn.linear_model import LinearRegression, Ridge
```

```python
def print_metric(y_true, y_pred):
    print(f'R^2: {r2_score(y_true, y_pred):.4}')
    print(f'MSE: {mean_squared_error(y_true, y_pred):.4}')
    print(f"Accuracy: {accuracy_score(y_true, y_pred):.4}")
    print(f"Precision: {precision_score(y_true, y_pred):.4}")
    print(f"Recall : {recall_score(y_true, y_pred):.4}")
    print(f"F-1 Score: {f1_score(y_true, y_pred):.4}")
```

First we split our data to train and test samples, with **0.7/0.3** portions.

```python

X_train, X_test, y_train, y_test = train_test_split(data[cols], data[target_col].values, test_size = 0.3, random_state= 142)
```

## Gradient Boosting Decision Trees


We cross validate X_train data with grid search cv, getting the best parameters for GBC

```python
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.model_selection import GridSearchCV


gdr_grid_search = GridSearchCV(GradientBoostingClassifier(),
                              [{"n_estimators": [30,80,150,450], "learning_rate": [0.01,0.05, 0.1, 0.15, 0.4]} ], 
                              cv = 5, error_score=make_scorer(accuracy_score), verbose=0)

gdr_grid_search.fit(X_train, y_train)



```

```python
print(gdr_grid_search.best_params_)
print(gdr_grid_search.best_score_)
print(gdr_grid_search.best_estimator_)
```

```python

best_gdr = GradientBoostingClassifier(learning_rate=0.4, n_estimators=30)
best_gdr.fit(X_train, y_train)
```

We get a metrics on test set, after this we can get most important features

```python
y_pred = best_gdr.predict(X_test)

print_metric(y_test, y_pred)
```

As we see here metrics are good so we can select features from this model.

### Feature importance Selection

We can get scores of features by build in method in Gradient Boosting Decision Trees.<br>


```python
# Most important features and their sorted indices
feature_importance_values = best_gdr.feature_importances_
sorted_idx = np.argsort(feature_importance_values)
```

```python
feature_importance_values.shape
```

```python
feature_importance_values[feature_importance_values==0].shape
```

Most of the importance of features are 0, so we can ignore them

```python
# get feature values which are non 0
feature_importance_values = feature_importance_values[feature_importance_values!=0]
sorted_idx = np.argsort(feature_importance_values)
```

In this step we get those important features as Series object

```python
# get columns of importantn features
important_features = data[cols].iloc[:,sorted_idx].columns
```

```python
# create a Series for important features and sort by feature importance
important_feature_series = pd.Series(feature_importance_values, index=important_features)
important_feature_series.sort_values(inplace=True)
```

Visualize most important feature values, we select **0.9** quartile as threshold. So the **10%** from the feature scores will be the most important features we want.

```python
# we copy our data remove indices for compact visualisation

feature_series_len = len(important_feature_series.index)
feature_series = important_feature_series.copy()
feature_series.index = [""]*feature_series_len

threshold = important_feature_series.quantile(0.9)

feature_series.plot.bar(figsize = (16, 4), label = "Features")
plt.hlines(threshold, -100, 10**3, color = "r", label = "threshold: 0.9 quartile")
plt.title("Feature Importance Scores")
plt.legend()
None
```

Then we can select our most important features from the Series, which values are higher then 90% percentile

```python
most_important_features = important_feature_series[important_feature_series> threshold]
```

```python
most_important_features.index.values[-2]
```

```python
# changing one index for it's shorter version
idx = most_important_features.index.values
idx[-2] = "Capital intensive business"
most_important_features.index = idx
```

```python
most_important_features
```

Visualise most important features

```python
most_important_features.plot.bar(figsize = (16,4))
plt.title("Most Important Feature's Scores")
None
```
